```python
import os

#读入文件
source_text=open('corpus.sentence.txt','r', encoding='utf-8')
      
wordlist=[]                                #词典
f=open('corpus.dict.txt','r', encoding='utf-8')
for eachline in f:
    nlen=len(eachline)-1;
    wordlist.append(eachline[:nlen])
f.close()
    
    
won=[]

for note in source_text:
    head=0
    i=10
    while (head <= len(note)):
        if (head>=(len(note)-i)):
            i=len(note)-head
        for p in range(i):
            rear=head+i-p
            flag=0
            for each in wordlist:
                if (note[head:rear]==each):
                    #print (each)
                    won.append(each)
                    head=head+len(each)
                    flag=1
                    break
            if (flag==1):
                break
        if (flag==0):
            won.append(note[head:head+1])
            head=head+1
    #won.append('\n')
            
file = open("corpus.output.txt", "a", encoding='utf-8')
for i in range(len(won)):
    s = str(won[i])
    if(s!="\n"):
        s = s+' '
    file.write(s)
file.close()

#print(won)
```

```python
import os

#读入文件
text1 = open('corpus.answer.txt','r', encoding='utf-8')
text2 = open("corpus.output.txt", 'r', encoding='utf-8')
file = open("corpus.judge.txt", "a", encoding='utf-8')

wordlist1 = []
wordlist2 = []
for eachline in text1:
    wordlist1.append(eachline)

for each in text2:
    wordlist2.append(each)

for i in range(len(wordlist1)):
    s1=str(wordlist1[i])
    #print(s1)
    s2=str(wordlist2[i])
    #print(s2)
    s1_list = s1.split()
    s2_list = s2.split()
    #print(s1_list[0])
    #print(s2_list[2])
    Correct=0
    for j in range(len(s2_list)):
        for k in range(j,len(s1_list)):
            if(s2_list[j]==s1_list[k]):
                Correct=Correct+1
                break
    Precision = Correct/len(s1_list)
    Recall = Correct/len(s2_list)
    F = (Precision) * (Recall) * 2 / (Precision + Recall)
    s = "Precision:"+str(Precision)+" "+"Recall:"+str(Recall)+" "+"F:"+str(F)+"Correct:"+str(Correct)+"l1:"+str(len(s1_list))+"l2:"+str(len(s2_list))+"\n"
    #print(Precision, end=' ')
    #print(Recall, end=' ')
    #print(F)
    file.write(s)
            
text1.close()
text2.close()
file.close()
```
